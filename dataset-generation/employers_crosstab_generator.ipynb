```python
fairYear = "2021"

df = pd.read_csv("~/Desktop/EuroClub/ECF_" + fairYear + "_Companies.csv", low_memory=False)
df.columns = ['Country', 'University']
print("Read the .CSV file.")
```

```python
# new columns for ID
df['id'] = '000'
```

```python
# countryDictionary = {
#     "AT": "Austria",
#     "BE": "Belgium",
#     "CH": "Switzerland",
#     "DE": "Germany",
#     "DK": "Denmark",
#     "ES": "Spain",
#     "FR": "France",
#     "FX": "Austria",
#     "GB": "Austria",
#     "NL": "Austria",
#     "US": "Austria"
# }

countryDictionary = {
    "Austria": "AT",
    "Belgium": "BE",
    "Switzerland": "CH",
    "Germany": "DE",
    "Denmark": "DK",
    "Italy": "IT",
    "Spain": "ES",
    "France": "FR",
    "Poland": "PL",
    # "Austria": "FX",
    "UK (w/o Scotland)": "GB",
    "Netherlands": "NL",
    "US": "US"
}
```

```python
replacementCount = 0

# for each country (row)
for i in range(len(df)):
    
    countryCode = countryDictionary[df["Country"][i]]
    print(countryCode)
    
    if(countryCode=="nan" or countryCode==""):
        print("Empty.")
        continue;
    
#     elif(countryCode=="GB"):
#         df["Country"][i] = "United Kingdom"
        
    elif(countryCode=="FX"):
        countryCode = "FR"
        

    try:
        countryInfo = countries.get(countryCode)
        print(countryInfo.name)

        df["Country"][i] = countryInfo.name
        df["id"][i] = countryInfo.numeric

    except:
        print("Error.")
```

# Make crosslist

```python
# Calculate number of records

colums2group = ['Country', 'id']
groupedEmployers = pd.DataFrame({'count': df.groupby(colums2group).size()})

#Source: https://stackoverflow.com/a/10374456

groupedEmployers
```

```python
groupedEmployers = groupedEmployers.rename(index={'ONE': 'one'})
```
