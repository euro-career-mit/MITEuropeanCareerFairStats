```python
def cat_json(outfile, infiles):
    
    file = open(outfile, "w")
    file.write("[%s]" % (",".join([mangle(file("FP_Groups_" + f + ".json").read()) for f in infiles])))

def mangle(s):
    return s.strip()[1:-1]
```
